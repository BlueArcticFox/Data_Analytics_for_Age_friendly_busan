```python
import folium
import pandas as pd
from folium.plugins import MarkerCluster 
#패키지 import
```

```python
df = pd.read_csv('./data/부산광역시_경로당현황_2021.csv'
, encoding='cp949') #부산 경로당 현황 읽어오기 
kizang_sub_df = df.loc[df['읍면동'].isin(['기장읍'])]#읍면동이 기장읍인 것만 자름
kizang_sub_df[['위도','경도','경로당명']].head(3)#출력 테스트
```

```python
jgwan_sub_df = df.loc[df['읍면동'].isin(['정관읍'])]#읍면동이 정관읍인 것만 자름
jgwan_sub_df[['위도','경도','경로당명']].head(3)
```

```python
kizang_pos= [35.23876594878238, 129.21575390899562] ## 기장읍 사무소
Jeonggwan_pos = [35.3231, 129.1784] # 정관읍 윗골공원
Noksan_pos= [35.08562495305441, 128.87931646088754]
Myungji_pos = [35.10820208025592, 128.9264932285998]

m = folium.Map(location=kizang_pos, zoom_start=17,width=750, 
               height=500)#지도 17의 줌으로 생성
coords = kizang_sub_df[['위도','경도','경로당명']]#자료에서 위도, 경도, 경로당명 가져와서
marker_cluster = MarkerCluster().add_to(m)#MarkerCluster을 지도에 추가하고
for lat,long,tool in zip(coords['위도'],coords['경도'],coords['경로당명']):#lat, long, tool을 이터레이터로 만들어서
    folium.Marker([lat,long],icon=folium.Icon(color='green'),tooltip=tool).add_to(marker_cluster)#초록색 마커로 마우스 갖다대면 이름 보이게 표시
m
```

```python
m_2 = folium.Map(location=Jeonggwan_pos, zoom_start=17,width=750, 
               height=500)#지도 17의 줌으로 생성
coords = jgwan_sub_df[['위도','경도','경로당명']]#자료에서 위도, 경도, 경로당명 가져와서
marker_cluster_2 = MarkerCluster().add_to(m_2)#MarkerCluster을 지도에 추가하고
for lat,long,tool in zip(coords['위도'],coords['경도'],coords['경로당명']):#lat, long, tool을 이터레이터로 만들어서
    folium.Marker([lat,long],icon=folium.Icon(color='green'),tooltip=tool).add_to(marker_cluster_2)#초록색 마커로 마우스 갖다대면 이름 보이게 표시
m_2
```
